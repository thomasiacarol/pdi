**GDAL**

```python
from osgeo import gdal
import numpy as np
```

```python
img2 = gdal.Open('/content/L71221071_07120010720_DN.tif')
```

```python
img3 = img2.ReadAsArray()
```

```python
img3.shape
```

```python
img3 = img3.swapaxes(0,1)
```

```python
img3.shape
```

```python
imshow(img3, (2,3,1))
```

```python
# Ler as bandas separadamente, transformar em array e depois empilhar cada uma
```

```python
b1 = img2.GetRasterBand(1).ReadAsArray()
b2 = img2.GetRasterBand(2).ReadAsArray()
b4 = img2.GetRasterBand(4).ReadAsArray()
```

```python
# Fazer empilhamento das bandas
stack = np.dstack([b1,b2,b4])
```

```python
stack.shape
```

```python
imshow(stack, (1,2,0))
```

```python
imshow(b1)
```
